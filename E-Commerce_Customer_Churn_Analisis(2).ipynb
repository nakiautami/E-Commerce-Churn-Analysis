## **1.1 Context**

**CONTEXT**

[E-commerce](https://www.ibm.com/id-id/topics/ecommerce?), atau perdagangan elektronik adalah proses pembelian dan penjualan barang atau jasa melalui media elektronik, khususnya internet. Dalam e-commerce, transaksi dilakukan secara online tanpa interaksi fisik antara penjual dan pembeli. Keberadaan platform digital mempermudah perusahaan dalam menjangkau pasar yang lebih luas dengan biaya yang lebih efisien. Namun, dengan pesatnya pertumbuhan e-commerce, persaingan semakin ketat, sehingga perusahaan menghadapi tantangan dalam meningkatkan serta mempertahankan pelanggan.

Pandemi COVID-19 pada tahun 2020 mendorong lonjakan transaksi e-commerce secara global. Pembatasan sosial membuat konsumen beralih ke belanja online untuk memenuhi kebutuhan sehari-hari. Data menunjukkan jumlah situs web e-commerce meningkat dari 17 juta pada 2020 menjadi 20 juta pada 2021, dan mencapai sekitar 30,7 juta pada pertengahan 2024 [(GlobeNewswire)](https://www.globenewswire.com/news-release/2024/12/30/3002507/28124/en/Growth-Trends-in-India-s-E-Commerce-Market-2024-2029-500-Million-Online-Shoppers-by-2029-India-Poised-to-Become-One-of-the-World-s-Largest-E-Commerce-Markets.html?) . Peningkatan jumlah platform e-commerce ini menandakan persaingan yang semakin ketat, menuntut perusahaan untuk menerapkan strategi berbasis data guna mempertahankan pelanggan.

Dalam menghadapi persaingan tersebut, perusahaan tidak hanya perlu mengumpulkan data, tetapi juga menganalisisnya secara mendalam untuk mengambil keputusan strategis yang lebih akurat. Salah satu metrik penting yang perlu diperhatikan adalah churn rate.

Churn adalah fenomena di mana pelanggan berhenti menggunakan layanan atau produk dalam jangka waktu tertentu sesuai dengan kriteria yang ditetapkan oleh perusahaan. Tingginya churn rate dapat berdampak negatif terhadap pendapatan berulang bulanan (MRR) serta mengindikasikan adanya ketidakpuasan pelanggan terhadap layanan yang diberikan. Oleh karena itu, mengukur dan menganalisis churn rate secara berkala menjadi penting bagi perusahaan untuk mengevaluasi efektivitas strategi retensi pelanggan dan meningkatkan kepuasan mereka [(Zendesk)](https://www.zendesk.com/blog/customer-churn-rate/#).

[digitalcommerce360](https://www.digitalcommerce360.com/article/us-ecommerce-sales/).Namun, salah satu tantangan utama bagi perusahaan adalah mempertahankan pelanggan agar tidak berpindah ke kompetitor. 3. Analisis Faktor-Faktor yang Mempengaruhi Churn

    - Faktor Demografi
        - Marital status: Status pernikahan dapat mempengaruhi kebiasaan belanja dan preferensi layanan
        - Gender: Perbedaan preferensi antara pelanggan pria dan wanita dalam memilih layanan

        - Number of device registered: Jumlah perangkat yang terdaftar pada akun pelanggan

    - Faktor Kepuasan dan Pengalaman Pengguna

        - Order amount like from last year: Perubahan jumlah pembelian dibandingkan tahun sebelumnya
        - Coupon used: Penggunaan kupon oleh pelanggan

        - Recency (R): Seberapa baru pelanggan melakukan transaksi terakhir

        - Monetary (M): Total nilai transaksi pelanggan dalam periode tertentu

    - Langkah-langkah:  Menghitung nilai RFM score untuk setiap pelanggan. Melakukan segmen pelanggan berdasarkan RFM score.

## **1.7 Evaluation Metric**

**A. Memahami Biaya Churn dalam E-Commerce**

Biaya churn dalam e-commerce mencakup dua komponen utama yang harus diperhatikan untuk menjaga profitabilitas bisnis:

1. **Biaya Akuisisi Pelanggan (Customer Acquisition Cost/CAC)**: adalah biaya yang dikeluarkan untuk mendapatkan pelanggan baru, termasuk pengeluaran untuk pemasaran dan promosi. Semakin tinggi CAC, semakin besar tantangan untuk memperoleh keuntungan dari pelanggan baru. Menurut [Shopify](https://www.shopify.com/blog/customer-acquisition-cost#), untuk menghitung CAC, jumlahkan semua biaya yang terkait dengan akuisisi pelanggan baru dalam periode tertentu, kemudian bagi totalnya dengan jumlah pelanggan baru yang diperoleh dalam periode tersebut.
Rumus CAC:
$$\text{CAC} = \frac{\text{Total biaya pemasaran dan penjualan dalam periode tertentu}}{\text{Jumlah pelanggan baru yang diperoleh dalam periode tersebut}}$$ 

2. **Biaya Retensi Pelanggan (Customer Retention Cost/CRC)**: Ini adalah biaya yang diperlukan untuk mempertahankan pelanggan yang sudah ada, termasuk program loyalitas, diskon, dan biaya layanan pelanggan. Investasi dalam retensi sering kali lebih efisien dibandingkan dengan akuisisi pelanggan baru. Menurut [Forbes](https://www.forbes.com/councils/forbesbusinesscouncil/2022/12/12/customer-retention-versus-customer-acquisition/), studi menunjukkan bahwa, tergantung pada industrinya, memperoleh pelanggan baru dapat memakan biaya lima hingga tujuh kali lebih besar daripada mempertahankan pelanggan lama.
Rumus CRC:
$$\text{CRC} = \frac{\text{Total retensi dalam periode tertentu}}{\text{Jumlah pelanggan yang dipertahankan dalam periode tersebut}}$$ 

Menurut [Forbes](https://www.forbes.com/councils/forbesbusinesscouncil/2022/12/12/customer-retention-versus-customer-acquisition/), **memperoleh pelanggan baru dapat memakan biaya lima hingga tujuh kali lebih besar** daripada mempertahankan pelanggan lama.


**B. Analisis Kesalahan dalam Prediksi Churn**

Dalam analisis churn pelanggan, terdapat empat kategori dalam confusion matrix:
- **True Positive (TP)** : Customer aktualnya churn dan diprediksi churn 
- **True Negative (TN)** : Customer aktualnya tidak churn dan diprediksi tidak churn
- **False Negative (FN)** : Customer aktualnya churn dan diprediksi tidak churn
- **False Positive (FP)** : Customer aktualnya tidak churn dan diprediksi churn

Dalam analisis prediksi churn pelanggan, penting untuk memahami dua jenis kesalahan yang dapat terjadi: Type 1 error (False Positive) dan Type 2 error (False Negative).

**Type 1 error** (False Positive): Kesalahan ini terjadi ketika pelanggan yang **sebenarnya tidak churn diprediksi akan churn**. Akibatnya, perusahaan mungkin memberikan insentif seperti diskon atau promosi kepada pelanggan yang sebenarnya tetap setia, yang dapat meningkatkan loyalitas tetapi juga menambah biaya operasional.

Keuntungan:
- Pelanggan setia merasa dihargai karena mendapatkan promo atau diskon, yang dapat meningkatkan reputasi e-commerce.

Kerugian:
- Biaya promosi meningkat karena diberikan kepada pelanggan yang tidak membutuhkannya.
- Sumber daya digunakan untuk strategi retensi yang kurang tepat sasaran.

**Type 2 error** (False Negative): Kesalahan ini terjadi ketika pelanggan yang **sebenarnya churn diprediksi tidak akan churn**. Hal ini dapat menyebabkan kerugian signifikan karena perusahaan kehilangan pelanggan tanpa upaya retensi. Selain itu, perusahaan harus mengeluarkan **Customer Acquisition Cost (CAC)**, yang membutuhkan biaya besar dan waktu lebih lama, karena menarik pelanggan baru lebih sulit dibandingkan mempertahankan pelanggan lama.

Kerugian:
- Kehilangan pelanggan yang benar-benar churn.
- Meningkatnya biaya untuk memperoleh pelanggan baru sebagai pengganti.

Dalam konteks analisis churn, **Type 2 Error lebih krusial untuk dihindari karena dampaknya yang lebih besar terhadap profitabilitas perusahaan**. Ketidakmampuan mengidentifikasi pelanggan yang berisiko churn dapat menyebabkan hilangnya pelanggan dan menignkatkan CAC. Sementara itu, meskipun **Type 1 Error juga perlu diperhatikan, efeknya cenderung lebih terkait pemborosan sumber daya tetapi tidak langsung mengurangi jumlah pelanggan**.

Untuk meminimalkan kedua jenis kesalahan ini, perusahaan dapat menggunakan analisis churn yang cermat dan strategi retensi yang tepat. Pendekatan ini membantu mengidentifikasi pelanggan yang berisiko churn dan memungkinkan perusahaan mengambil tindakan proaktif untuk mempertahankan mereka, seperti menawarkan insentif yang relevan atau meningkatkan layanan pelanggan.

Selain itu, penting bagi perusahaan untuk terus memantau dan menganalisis data pelanggan guna memahami pola perilaku yang mengarah pada churn. Dengan demikian, perusahaan dapat mengembangkan strategi yang lebih efektif dalam mengurangi tingkat churn dan meningkatkan loyalitas pelanggan.

**C. Perbandingan Biaya akuisisi pelanggan (Customer Acquisition Cost/CAC)** dan **biaya retensi pelanggan (Cost Retention)**.

**1.Biaya Akuisisi Pelanggan (CAC)**: 
- Total biaya pemasaran dan penjualan untuk mendapatkan pelanggan baru.
- Contoh: Jika perusahaan menghabiskan $90.000 untuk mendapatkan 1.000 pelanggan dalam setahun, maka CAC = $90 per pelanggan.[(IBM)](https://www.ibm.com/id-id/think/topics/customer-acquisition?).

**2. Cost Retention**:
- Total biaya yang dikeluarkan untuk mempertahankan pelanggan yang sudah ada.

- Contoh: Jika perusahaan menghabiskan $100.000 untuk mempertahankan 2.500 pelanggan dalam setahun, maka CRC = $40 per pelanggan.[(ottopoint)](https://www.ottopoint.id/id/article/akuisisi-vs-retensi-pada-bisnis-mana-lebih-baik?)

Karena biaya retensi umumnya lebih rendah daripada biaya akuisisi, perusahaan perlu mengoptimalkan strategi retensi pelanggan agar dapat mengurangi churn dan meningkatkan profitabilitas jangka panjang.

**D. Mengoptimalkan Pengelolaan Churn dengan F2-Score**

Untuk mencapai keseimbangan dalam pengelolaan churn, perusahaan dapat memanfaatkan metrik F2-Score. F2-Score adalah varian dari F1-Score yang menggabungkan precision dan recall menjadi satu nilai, dengan penekanan lebih besar pada recall. Dalam analisis churn, penggunaan F2-Score sangat penting karena:

- Fokus pada Pelanggan Berisiko Churn
F2-Score lebih mengutamakan recall, sehingga perusahaan dapat lebih mudah menemukan pelanggan yang kemungkinan besar akan berhenti menggunakan layanan. Hal ini sangat penting bagi perusahaan yang ingin mempertahankan lebih banyak pelanggan dan mengurangi tingkat churn.

- Optimalisasi Sumber Daya
Dengan memahami faktor penyebab churn, perusahaan dapat mengalokasikan sumber daya secara lebih efisien dan menghindari pemborosan pada pelanggan yang tetap loyal.

- Menyeimbangkan Kesalahan Prediksi
Metrik ini membantu menyeimbangkan antara false positives (pelanggan yang diprediksi churn tetapi sebenarnya tidak) dan false negatives (pelanggan yang tidak terdeteksi akan churn tetapi sebenarnya pergi). Dengan pendekatan ini, strategi retensi dapat lebih efektif.

Rumus F2-Score:
$$F_2= \frac{(1+2^2) \times \text{Precision} \times \text{Recall}}{2^2 \times \text{Precision} \times \text{Recall}}$$

Nilai F2-Score berkisar antara 0 hingga 1, di mana nilai yang lebih tinggi menunjukkan kinerja prediksi yang lebih baik. Dengan menggunakan F2-Score, perusahaan dapat lebih memprioritaskan pencegahan churn tanpa mengabaikan akurasi prediksi.

Ada [tiga jenis utama missing value](https://www.geeksforgeeks.org/ml-handling-missing-values/) dalam data:

1. Hilang Secara Acak (MCAR)
Data hilang secara sepenuhnya acak dan tidak dipengaruhi oleh variabel lain dalam dataset. Dengan kata lain, hilangnya data tidak berkaitan dengan nilai variabel lain maupun karakteristik data itu sendiri.

2. Hilang Secara Acak (MAR)
Kehilangan data dipengaruhi oleh variabel lain dalam dataset, tetapi tidak oleh variabel yang hilang itu sendiri. Artinya, ada pola tertentu dalam hilangnya data yang dapat diprediksi berdasarkan informasi yang tersedia.

3. Hilang Tidak Secara Acak (MNAR)
Kehilangan data terjadi karena alasan yang berkaitan langsung dengan nilai yang hilang itu sendiri. Dengan kata lain, data hilang bukan secara kebetulan, tetapi karena ada faktor tertentu yang menyebabkan nilai tersebut tidak dilaporkan atau tidak tersedia.


Berdasarkan heatmap **jenis missing value Missing data ini ialah Completely at Random (MCAR)**.
- MCAR terjadi jika nilai yang hilang tersebar secara acak tanpa pola tertentu dan tidak berhubungan dengan variabel lain.
- Dari heatmap ini, korelasi antar kolom yang memiliki missing values sangat lemah (-0.1). Ini menunjukkan bahwa beberapa kolom bisa memiliki missing values secara acak.

## **3.5 Pengecekan Outlier**

```python
# Daftar kolom numerik yang di cek
kolom = [
    'Tenure', 'CityTier', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered',
    'SatisfactionScore', 'NumberOfAddress', 'Complain', 'OrderAmountHikeFromlastYear',
    'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CashbackAmount'
]

# Inisialisasi list untuk menyimpan hasil deteksi outlier
result = []

# Pengecekan outlier menggunakan metode IQR
for col in kolom:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    lower_outliers = df[df[col] < lower_limit][col]
    upper_outliers = df[df[col] > upper_limit][col]
    
    total_outliers = len(lower_outliers) + len(upper_outliers)
    
    if total_outliers > 0:
        result.append({
            'Kolom': col,
            'Jumlah outlier lower': len(lower_outliers),
            'Jumlah outlier upper': len(upper_outliers),
            'Total outlier (%)': f'{(total_outliers / len(df)) * 100:.2f}%',
            'Nilai outlier lower (<=)': lower_outliers.max() if len(lower_outliers) > 0 else 'None',
            'Nilai outlier upper (>=)': upper_outliers.min() if len(upper_outliers) > 0 else 'None'
        })

# Konversi hasil deteksi outlier ke DataFrame
outlier_df = pd.DataFrame(result)

# Menampilkan tabel outlier
print("Tabel daftar outlier:")
outlier_df
```

## **4.1.1 Statistik Deskriptif**

Analisis deskriptif dan uji normalitas merupakan langkah awal penting dalam eksplorasi data untuk memahami distribusi, pola, dan potensi outlier.Statistik Deskriptif adalah salah satu langkah utama dalam Exploratory Data **Analysis (EDA) yang digunakan untuk merangkum dan memahami karakteristik data secara kuantitatif. Statistik ini membantu dalam memberikan gambaran umum mengenai distribusi data sebelum dilakukan analisis lebih lanjut**. Statistik deskriptif mencakup perhitungan mean, median, standar deviasi, serta nilai minimum dan maksimum guna membantu pemilihan metode analisis yang sesuai (parametrik atau non-parametrik).

Uji normalitas, seperti Shapiro-Wilk test, digunakan untuk menentukan apakah data berdistribusi normal, yang menjadi dasar bagi banyak metode statistik seperti regresi linear dan uji t. Jika data tidak normal (p-value ≤ 0.05), maka metode alternatif seperti Mann-Whitney U Test atau Kruskal-Wallis Test lebih disarankan. Analisis ini membantu dalam pengambilan keputusan dan pemilihan teknik pemodelan yang lebih akurat.

```python
# Daftar kolom numerik yang akan dianalisis
num_columns = [
    'Tenure', 'CityTier', 'WarehouseToHome', 'HourSpendOnApp',
    'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress',
    'Complain', 'OrderAmountHikeFromLastYear', 'CouponUsed', 'OrderCount',
    'DaySinceLastOrder', 'CashbackAmount'
]

# Membuat list untuk menyimpan ringkasan statistik deskriptif dan uji normalitas
summary_list = [
    {
        'Column': col,
        'Mean': df[col].mean(),
        'Median': df[col].median(),
        'Std Dev': df[col].std(),
        'Min': df[col].min(),
        'Max': df[col].max(),
        'p-value': f'{shapiro(df[col])[1]:.5f}',
        'Normalitas': 'Distribusi Normal' if shapiro(df[col])[1] > 0.05 else 'Distribusi Tidak Normal'
    }
    for col in num_columns
]

summary_df = pd.DataFrame(summary_list)

# Menampilkan hasil ringkasan statistik deskriptif dan uji normalitas
summary_df
```

# **4.2 Data Correlation**

Analisis korelasi bertujuan untuk memahami hubungan antar variabel dalam dataset dengan menghitung seberapa kuat keterkaitan antara dua variabel. Salah satu metode yang digunakan adalah **Spearman’s Rank Correlation**, yang lebih sesuai untuk data dengan hubungan non-linear atau distribusi tidak normal. Korelasi berkisar antara -1 hingga 1, di mana nilai 1 menunjukkan hubungan positif sempurna, -1 menunjukkan hubungan negatif sempurna, dan 0 berarti tidak ada korelasi. Untuk memvisualisasikan hubungan ini, heatmap korelasi digunakan dengan skala warna coolwarm, di mana warna merah menunjukkan korelasi positif yang kuat, biru menunjukkan korelasi negatif yang kuat, dan warna netral menunjukkan korelasi lemah.

Derajat Hubungan antara Variabel : interpretasi koefisien korelasi

| Value of Correlation Coeffficient | Relationship Interpretation |
| --- | --- |
| 0.0 - 0.2 | Very Weak Correlation|
| 0.2 - 0.4 | Weak Correlation|
| 0.4 - 0.6 | Moderate Correlation|
| 0.6 - 0.9 | Strong Correlation|
| 0.9 - 1.0 | Very Strong Correlation|

```python
# Membuat figure dan axis dengan ukuran yang lebih besar untuk visualisasi yang lebih jelas
fig, ax = plt.subplots(figsize=(20, 16))

# Menghitung matriks korelasi menggunakan metode Spearman
# Metode ini digunakan karena lebih sesuai untuk data dengan hubungan non-linear atau tidak berdistribusi normal
corr_matrix = df.corr(method='spearman', numeric_only=True)
mask = np.triu(np.ones_like(corr_matrix))

# Membuat heatmap menggunakan Seaborn
sns.heatmap(
    corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0, mask=mask, 
    square=True, linewidths=0.5, annot_kws={"size": 10}, ax=ax
)

plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
def create_plot(data, col, x_maks=2000):
    """
    Membuat dua visualisasi terkait churn berdasarkan kategori tertentu:
    1. Grafik batang yang menunjukkan persentase churn berdasarkan kategori.
    2. Grafik batang yang menunjukkan jumlah pelanggan berdasarkan kategori dan status churn.

    Parameters:
    data (DataFrame): Dataset yang berisi informasi pelanggan dan churn.
    col (str): Nama kolom kategori yang akan dianalisis.
    x_maks (int, optional): Batas maksimum sumbu-x untuk grafik jumlah pelanggan. Default adalah 2000.
    """
    # Menghitung jumlah churn berdasarkan kategori
    churn_summary = data.groupby([col, 'Churn']).size().unstack(fill_value=0).reset_index()
    churn_summary.columns = [col, 'Churned', 'Not Churned']
    churn_summary['Total_Customers'] = churn_summary['Churned'] + churn_summary['Not Churned']
    churn_summary['Churn_Rate (%)'] = (churn_summary['Churned'] / churn_summary['Total_Customers'] * 100).round(2)
    churn_summary = churn_summary.sort_values(by='Churn_Rate (%)', ascending=False)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    
    # Grafik 1: Persentase Churn
    grafik1 = sns.barplot(data=churn_summary, y=col, x='Churn_Rate (%)', ax=axes[0], color='#3498db')
    labels = ['{:.1f} %'.format(x) for x in churn_summary['Churn_Rate (%)']]
    grafik1.bar_label(grafik1.containers[0], labels=labels, padding=3, label_type='center')
    grafik1.set_title(f'Persentase Churn berdasarkan {col}', fontsize=12, fontweight='bold')
    grafik1.set_xlabel('% Churn')
    grafik1.set_ylabel(col)
    grafik1.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    churn_rate = data['Churn'].value_counts(normalize=True).get(1, 0) * 100
    grafik1.axvline(x=churn_rate, color='red', linestyle='--', label=f'Rata-rata Churn: {churn_rate:.2f}%')
    grafik1.legend(loc='lower right')
    
    # Grafik 2: Jumlah pelanggan berdasarkan kategori
    churn_summary = churn_summary.sort_values(by='Total_Customers', ascending=False)
    grafik2 = sns.countplot(data=data, y=col, hue='Churn', order=churn_summary[col], palette=['#e74c3c', '#2ecc71'], ax=axes[1])
    grafik2.set_title(f'Jumlah Pelanggan berdasarkan {col} dan Churn', fontsize=12, fontweight='bold')
    for container in grafik2.containers:
        grafik2.bar_label(container)
    grafik2.set_xlim(0, x_maks)
    grafik2.set_xlabel('Jumlah')
    grafik2.set_ylabel(col)
    grafik2.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    
    plt.tight_layout()
    plt.show()

# Pemanggilan fungsi dengan kolom "MaritalStatus"
create_plot(df, 'MaritalStatus', x_maks=6000)
```

```python
# Membuat tabel kontingensi menggunakan crosstab
contingency_table = pd.crosstab(df['Churn'], df['MaritalStatus'])

# Nilai signifikansi
alpha = 0.05

# Uji Chi-Square
stat, pvalue, dof, expected_freq = chi2_contingency(contingency_table)

# Interpretasi hasil
if pvalue > alpha:
    print(f'P-Value: {pvalue:.5f} > {alpha}. Fail to Reject Ho.')
    print('Kita tidak punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara MaritalStatus dan Churn.')
else:
    print(f'P-Value: {pvalue:.5f} <= {alpha}. Reject Ho.')
    print('Kita punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara MaritalStatus dan Churn.')
```

```python
def create_plot(data, col, x_maks=2000):
    """
    Membuat dua visualisasi terkait churn berdasarkan kategori tertentu:
    1. Grafik batang yang menunjukkan persentase churn berdasarkan kategori.
    2. Grafik batang yang menunjukkan jumlah pelanggan berdasarkan kategori dan status churn.

    Parameters:
    data (DataFrame): Dataset yang berisi informasi pelanggan dan churn.
    col (str): Nama kolom kategori yang akan dianalisis.
    x_maks (int, optional): Batas maksimum sumbu-x untuk grafik jumlah pelanggan. Default adalah 2000.
    """
    # Menghitung jumlah churn berdasarkan kategori
    churn_summary = data.groupby([col, 'Churn']).size().unstack(fill_value=0).reset_index()
    churn_summary.columns = [col, 'Churned', 'Not Churned']
    churn_summary['Total_Customers'] = churn_summary['Churned'] + churn_summary['Not Churned']
    churn_summary['Churn_Rate (%)'] = (churn_summary['Churned'] / churn_summary['Total_Customers'] * 100).round(2)
    churn_summary = churn_summary.sort_values(by='Churn_Rate (%)', ascending=False)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    
    # Grafik 1: Persentase Churn
    grafik1 = sns.barplot(data=churn_summary, y=col, x='Churn_Rate (%)', ax=axes[0], color='#3498db')
    labels = ['{:.1f} %'.format(x) for x in churn_summary['Churn_Rate (%)']]
    grafik1.bar_label(grafik1.containers[0], labels=labels, padding=3, label_type='center')
    grafik1.set_title(f'Persentase Churn berdasarkan {col}', fontsize=12, fontweight='bold')
    grafik1.set_xlabel('% Churn')
    grafik1.set_ylabel(col)
    grafik1.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    churn_rate = data['Churn'].value_counts(normalize=True).get(1, 0) * 100
    grafik1.axvline(x=churn_rate, color='red', linestyle='--', label=f'Rata-rata Churn: {churn_rate:.2f}%')
    grafik1.legend(loc='lower right')
    
    # Grafik 2: Jumlah pelanggan berdasarkan kategori
    churn_summary = churn_summary.sort_values(by='Total_Customers', ascending=False)
    grafik2 = sns.countplot(data=data, y=col, hue='Churn', order=churn_summary[col], palette=['#e74c3c', '#2ecc71'], ax=axes[1])
    grafik2.set_title(f'Jumlah Pelanggan berdasarkan {col} dan Churn', fontsize=12, fontweight='bold')
    for container in grafik2.containers:
        grafik2.bar_label(container)
    grafik2.set_xlim(0, x_maks)
    grafik2.set_xlabel('Jumlah')
    grafik2.set_ylabel(col)
    grafik2.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    
    plt.tight_layout()
    plt.show()

# Pemanggilan fungsi dengan kolom "CityTier"
create_plot(df, 'CityTier', x_maks=6000)
```

```python
# Membuat tabel kontingensi menggunakan crosstab
contingency_table = pd.crosstab(df['Churn'], df['CityTier'])

# Nilai signifikansi
alpha = 0.05

# Uji Chi-Square
stat, pvalue, dof, expected_freq = chi2_contingency(contingency_table)

# Interpretasi hasil
if pvalue > alpha:
    print(f'P-Value: {pvalue:.5f} > {alpha}. Fail to Reject Ho.')
    print('Kita tidak punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara CityTier dan Churn.')
else:
    print(f'P-Value: {pvalue:.5f} <= {alpha}. Reject Ho.')
    print('Kita punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara CityTier dan Churn.')
```

```python
from scipy.stats import mannwhitneyu

# Menentukan tingkat signifikansi (alpha)
alpha = 0.05

# Memisahkan data tenure berdasarkan status churn
tenure_churned = df[df['Churn'] == 'Churned']['Tenure']
tenure_not_churned = df[df['Churn'] == 'Not Churned']['Tenure']

# Melakukan Uji Mann-Whitney untuk membandingkan distribusi tenure antara dua kelompok
stat, pvalue = mannwhitneyu(
    x=tenure_churned, 
    y=tenure_not_churned, 
    alternative='two-sided'  # Menggunakan pengujian dua sisi
)

# Menentukan kesimpulan berdasarkan nilai p-value
if pvalue > alpha:
    print(f'P-Value: {pvalue:.5f} > {alpha}. Fail to Reject Ho. Median tenure pada konsumen churn dan tidak churn adalah sama.')
else:
    print(f"P-Value: {pvalue:.5f} <= {alpha}. Reject Ho. Minimal ada 1 perbedaan signifikan antara median tenure berdasarkan churn.")
```

```python
# Menentukan tingkat signifikansi (alpha)
alpha = 0.05

# Memisahkan data NumberOfDeviceRegistered berdasarkan status churn
NumberOfDeviceRegistered_churned = df[df['Churn'] == 'Churned']['NumberOfDeviceRegistered']
NumberOfDeviceRegistered_not_churned = df[df['Churn'] == 'Not Churned']['NumberOfDeviceRegistered']

# Melakukan Uji Mann-Whitney untuk membandingkan distribusi NumberOfDeviceRegistered antara dua kelompok
stat, pvalue = mannwhitneyu(
    x=NumberOfDeviceRegistered_churned, 
    y=NumberOfDeviceRegistered_not_churned, 
    alternative='two-sided'  # Menggunakan pengujian dua sisi
)

# Menentukan kesimpulan berdasarkan nilai p-value
if pvalue > alpha:
    print(f'P-Value: {pvalue:.5f} > {alpha}. Fail to Reject Ho. Median NumberOfDeviceRegistered pada konsumen churn dan tidak churn adalah sama.')
else:
    print(f"P-Value: {pvalue:.5f} <= {alpha}. Reject Ho. Minimal ada 1 perbedaan signifikan antara median NumberOfDeviceRegistered berdasarkan churn.")
```

```python
def create_plot(data, col, x_maks=2000):
    """
    Membuat dua visualisasi terkait churn berdasarkan kategori tertentu:
    1. Grafik batang yang menunjukkan persentase churn berdasarkan kategori.
    2. Grafik batang yang menunjukkan jumlah pelanggan berdasarkan kategori dan status churn.

    Parameters:
    data (DataFrame): Dataset yang berisi informasi pelanggan dan churn.
    col (str): Nama kolom kategori yang akan dianalisis.
    x_maks (int, optional): Batas maksimum sumbu-x untuk grafik jumlah pelanggan. Default adalah 2000.
    """
    # Menghitung jumlah churn berdasarkan kategori
    churn_summary = data.groupby([col, 'Churn']).size().unstack(fill_value=0).reset_index()
    churn_summary.columns = [col, 'Churned', 'Not Churned']
    churn_summary['Total_Customers'] = churn_summary['Churned'] + churn_summary['Not Churned']
    churn_summary['Churn_Rate (%)'] = (churn_summary['Churned'] / churn_summary['Total_Customers'] * 100).round(2)
    churn_summary = churn_summary.sort_values(by='Churn_Rate (%)', ascending=False)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    
    # Grafik 1: Persentase Churn
    grafik1 = sns.barplot(data=churn_summary, y=col, x='Churn_Rate (%)', ax=axes[0], color='#3498db')
    labels = ['{:.1f} %'.format(x) for x in churn_summary['Churn_Rate (%)']]
    grafik1.bar_label(grafik1.containers[0], labels=labels, padding=3, label_type='center')
    grafik1.set_title(f'Persentase Churn berdasarkan {col}', fontsize=12, fontweight='bold')
    grafik1.set_xlabel('% Churn')
    grafik1.set_ylabel(col)
    grafik1.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    churn_rate = data['Churn'].value_counts(normalize=True).get(1, 0) * 100
    grafik1.axvline(x=churn_rate, color='red', linestyle='--', label=f'Rata-rata Churn: {churn_rate:.2f}%')
    grafik1.legend(loc='lower right')
    
    # Grafik 2: Jumlah pelanggan berdasarkan kategori
    churn_summary = churn_summary.sort_values(by='Total_Customers', ascending=False)
    grafik2 = sns.countplot(data=data, y=col, hue='Churn', order=churn_summary[col], palette=['#e74c3c', '#2ecc71'], ax=axes[1])
    grafik2.set_title(f'Jumlah Pelanggan berdasarkan {col} dan Churn', fontsize=12, fontweight='bold')
    for container in grafik2.containers:
        grafik2.bar_label(container)
    grafik2.set_xlim(0, x_maks)
    grafik2.set_xlabel('Jumlah')
    grafik2.set_ylabel(col)
    grafik2.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    
    plt.tight_layout()
    plt.show()

# Pemanggilan fungsi dengan kolom "Complain"
create_plot(df, 'Complain', x_maks=6000)
```

```python
# Membuat tabel kontingensi menggunakan crosstab
contingency_table = pd.crosstab(df['Churn'], df['Complain'])

# Nilai signifikansi
alpha = 0.05

# Uji Chi-Square
stat, pvalue, dof, expected_freq = chi2_contingency(contingency_table)

# Interpretasi hasil
if pvalue > alpha:
    print(f'P-Value: {pvalue:.5f} > {alpha}. Fail to Reject Ho.')
    print('Kita tidak punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara Complain dan Churn.')
else:
    print(f'P-Value: {pvalue:.5f} <= {alpha}. Reject Ho.')
    print('Kita punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara Complain dan Churn.')
```

```python
def create_plot(data, col, x_maks=2000):
    """
    Membuat dua visualisasi terkait churn berdasarkan kategori tertentu:
    1. Grafik batang yang menunjukkan persentase churn berdasarkan kategori.
    2. Grafik batang yang menunjukkan jumlah pelanggan berdasarkan kategori dan status churn.

    Parameters:
    data (DataFrame): Dataset yang berisi informasi pelanggan dan churn.
    col (str): Nama kolom kategori yang akan dianalisis.
    x_maks (int, optional): Batas maksimum sumbu-x untuk grafik jumlah pelanggan. Default adalah 2000.
    """
    # Menghitung jumlah churn berdasarkan kategori
    churn_summary = data.groupby([col, 'Churn']).size().unstack(fill_value=0).reset_index()
    churn_summary.columns = [col, 'Churned', 'Not Churned']
    churn_summary['Total_Customers'] = churn_summary['Churned'] + churn_summary['Not Churned']
    churn_summary['Churn_Rate (%)'] = (churn_summary['Churned'] / churn_summary['Total_Customers'] * 100).round(2)
    churn_summary = churn_summary.sort_values(by='Churn_Rate (%)', ascending=False)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    
    # Grafik 1: Persentase Churn
    grafik1 = sns.barplot(data=churn_summary, y=col, x='Churn_Rate (%)', ax=axes[0], color='#3498db')
    labels = ['{:.1f} %'.format(x) for x in churn_summary['Churn_Rate (%)']]
    grafik1.bar_label(grafik1.containers[0], labels=labels, padding=3, label_type='center')
    grafik1.set_title(f'Persentase Churn berdasarkan {col}', fontsize=12, fontweight='bold')
    grafik1.set_xlabel('% Churn')
    grafik1.set_ylabel(col)
    grafik1.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    churn_rate = data['Churn'].value_counts(normalize=True).get(1, 0) * 100
    grafik1.axvline(x=churn_rate, color='red', linestyle='--', label=f'Rata-rata Churn: {churn_rate:.2f}%')
    grafik1.legend(loc='lower right')
    
    # Grafik 2: Jumlah pelanggan berdasarkan kategori
    churn_summary = churn_summary.sort_values(by='Total_Customers', ascending=False)
    grafik2 = sns.countplot(data=data, y=col, hue='Churn', order=churn_summary[col], palette=['#e74c3c', '#2ecc71'], ax=axes[1])
    grafik2.set_title(f'Jumlah Pelanggan berdasarkan {col} dan Churn', fontsize=12, fontweight='bold')
    for container in grafik2.containers:
        grafik2.bar_label(container)
    grafik2.set_xlim(0, x_maks)
    grafik2.set_xlabel('Jumlah')
    grafik2.set_ylabel(col)
    grafik2.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    
    plt.tight_layout()
    plt.show()

# Pemanggilan fungsi dengan kolom "PreferredOrderCat"
create_plot(df, 'PreferredOrderCat', x_maks=6000)
```

```python
def create_plot(data, col, x_maks=2000):
    """
    Membuat dua visualisasi terkait churn berdasarkan kategori tertentu:
    1. Grafik batang yang menunjukkan persentase churn berdasarkan kategori.
    2. Grafik batang yang menunjukkan jumlah pelanggan berdasarkan kategori dan status churn.

    Parameters:
    data (DataFrame): Dataset yang berisi informasi pelanggan dan churn.
    col (str): Nama kolom kategori yang akan dianalisis.
    x_maks (int, optional): Batas maksimum sumbu-x untuk grafik jumlah pelanggan. Default adalah 2000.
    """
    # Menghitung jumlah churn berdasarkan kategori
    churn_summary = data.groupby([col, 'Churn']).size().unstack(fill_value=0).reset_index()
    churn_summary.columns = [col, 'Churned', 'Not Churned']
    churn_summary['Total_Customers'] = churn_summary['Churned'] + churn_summary['Not Churned']
    churn_summary['Churn_Rate (%)'] = (churn_summary['Churned'] / churn_summary['Total_Customers'] * 100).round(2)
    churn_summary = churn_summary.sort_values(by='Churn_Rate (%)', ascending=False)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    
    # Grafik 1: Persentase Churn
    grafik1 = sns.barplot(data=churn_summary, y=col, x='Churn_Rate (%)', ax=axes[0], color='#3498db')
    labels = ['{:.1f} %'.format(x) for x in churn_summary['Churn_Rate (%)']]
    grafik1.bar_label(grafik1.containers[0], labels=labels, padding=3, label_type='center')
    grafik1.set_title(f'Persentase Churn berdasarkan {col}', fontsize=12, fontweight='bold')
    grafik1.set_xlabel('% Churn')
    grafik1.set_ylabel(col)
    grafik1.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    churn_rate = data['Churn'].value_counts(normalize=True).get(1, 0) * 100
    grafik1.axvline(x=churn_rate, color='red', linestyle='--', label=f'Rata-rata Churn: {churn_rate:.2f}%')
    grafik1.legend(loc='lower right')
    
    # Grafik 2: Jumlah pelanggan berdasarkan kategori
    churn_summary = churn_summary.sort_values(by='Total_Customers', ascending=False)
    grafik2 = sns.countplot(data=data, y=col, hue='Churn', order=churn_summary[col], palette=['#e74c3c', '#2ecc71'], ax=axes[1])
    grafik2.set_title(f'Jumlah Pelanggan berdasarkan {col} dan Churn', fontsize=12, fontweight='bold')
    for container in grafik2.containers:
        grafik2.bar_label(container)
    grafik2.set_xlim(0, x_maks)
    grafik2.set_xlabel('Jumlah')
    grafik2.set_ylabel(col)
    grafik2.grid(axis='x', linestyle='--', linewidth=0.5, color='gray')
    
    plt.tight_layout()
    plt.show()

# Pemanggilan fungsi dengan kolom "PreferredPaymentMode"
create_plot(df, 'PreferredPaymentMode', x_maks=6000)
```

```python
# Membuat tabel kontingensi menggunakan crosstab
contingency_table = pd.crosstab(df['Churn'], df['PreferredPaymentMode'])

# Nilai signifikansi
alpha = 0.05

# Uji Chi-Square
stat, pvalue, dof, expected_freq = chi2_contingency(contingency_table)

# Interpretasi hasil
if pvalue > alpha:
    print(f'P-Value: {pvalue:.5f} > {alpha}. Fail to Reject Ho.')
    print('Kita tidak punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara PreferredPaymentMode dan Churn.')
else:
    print(f'P-Value: {pvalue:.5f} <= {alpha}. Reject Ho.')
    print('Kita punya cukup bukti untuk menyatakan bahwa terdapat keterkaitan antara PreferredPaymentMode dan Churn.')
```

## **4.5 Analisis RFM (Regency, Frequency, Monetary) untuk Churn**

RFM Analysis adalah metode segmentasi pelanggan berdasarkan **Recency (seberapa baru transaksi terakhir), Frequency (seberapa sering pelanggan bertransaksi), dan Monetary (berapa banyak uang yang dihabiskan pelangggan)**.
RFM digunakan untuk memahami perilaku pelanggan. Analisis ini membantu dalam:

- Mengidentifikasi pelanggan loyal dan potensial churn – Pelanggan dengan skor Recency rendah atau Frequency yang jarang bertransaksi cenderung memiliki risiko churn lebih tinggi.
- Memudahkan segmentasi awal – RFM membantu membagi pelanggan ke dalam kategori seperti Best Customers, Loyal Customers, At Risk, dan Churned Customers, yang dapat digunakan sebagai dasar dalam pengembangan model Machine Learning.
- Menentukan strategi pemasaran yang lebih efektif – Dengan RFM, kita bisa menargetkan pelanggan dengan strategi berbeda berdasarkan segmen mereka sebelum menerapkan model prediksi churn yang lebih kompleks.

Mengubah nama kolom RFM:
- `DaySinceLastOrder` -> Regency (seberapa lama sejak transaksi terakhir)
- `OrderCount` -> Frequency (seberapa sering pelanggan bertransaksi)
- `CashbackAmount` -> Monetary (nilai total cashback yang diterima sebagai representasi pengeluaran pelanggan)

```python
def compute_rfm(df):
    rfm_df = df.copy()
    rfm_df.rename(columns={'DaySinceLastOrder': 'Recency', 
                           'OrderCount': 'Frequency', 
                           'CashbackAmount': 'Monetary'}, inplace=True)

    bins_recency = [-1, 7, 14, 30, 31]
    bins_frequency = [0, 3, 7, 12, 16]
    bins_monetary = [0, 50, 150, 250, 325]
    labels = [4, 3, 2, 1]

    rfm_df['R_Score'] = pd.cut(rfm_df['Recency'], bins=bins_recency, labels=labels, include_lowest=True).astype('int64')
    rfm_df['F_Score'] = pd.cut(rfm_df['Frequency'], bins=bins_frequency, labels=labels, include_lowest=True).astype('int64')
    rfm_df['M_Score'] = pd.cut(rfm_df['Monetary'], bins=bins_monetary, labels=labels, include_lowest=True).astype('int64')

    rfm_df['RFM_Score'] = rfm_df['R_Score'] + rfm_df['F_Score'] + rfm_df['M_Score']

    def segment_customer(score):
        if score >= 10:
            return "Best Customers"
        elif score >= 8:
            return "Loyal Customers"
        elif score >= 6:
            return "Potential Loyalists"
        elif score >= 4:
            return "Need attention"
        else:
            return "At Risk"

    rfm_df['Customer_Segment'] = rfm_df['RFM_Score'].apply(segment_customer)
    
    return rfm_df
```

Berdasarkan analisis data, churn dalam perusahaan ini tidak hanya dipengaruhi oleh faktor demografi, tetapi juga oleh perilaku pengguna, tingkat kepuasan, pengalaman, dan pola transaksi. Faktor-faktor ini berperan penting dalam mencegah pelanggan berhenti bertransaksi.

Secara umum, indikator churn dalam e-commerce mencakup:

- **Tidak bertransaksi dalam periode tertentu (misalnya, 1 atau 3 bulan terakhir)**.
    -Pelanggan yang mulai pasif masuk ke segmen Classic, sedangkan yang sudah lama tidak bertransaksi masuk ke segmen Churn.

- **Penurunan frekuensi pembelian dibandingkan periode sebelumnya**.
    - Analisis RFM (Recency, Frequency, Monetary) digunakan untuk melihat perubahan pola transaksi pelanggan.
    - Jika frekuensi pembelian berkurang, pelanggan dapat turun level.

- **Penurunan engagement (jarang membuka aplikasi, tidak menggunakan kupon, atau tidak mengklik promo)**.
    - Pelanggan di segmen Silver membutuhkan lebih banyak engagement agar tetap aktif.
    - Pelanggan di segmen Classic perlu didorong dengan promo dan survei kepuasan untuk meningkatkan interaksi mereka.

- **Keluar dari program loyalitas atau tidak menggunakan reward yang tersedia.**
    - Jika pelanggan Platinum atau Gold mulai jarang memanfaatkan reward, mereka bisa turun ke segmen Classic, yang menandakan risiko churn lebih tinggi.

Setelah model memprediksi customer churn, perusahaan e-commerce harus menyesuaikan strategi retensi berdasarkan segmentasi pelanggan agar alokasi sumber daya lebih efektif. Dengan keterbatasan anggaran pemasaran, perusahaan perlu memprioritaskan pelanggan dengan kontribusi pendapatan terbesar untuk memaksimalkan Customer Lifetime Value (CLV).
- Best Customers menjadi fokus utama karena memiliki nilai transaksi tinggi dan loyalitas kuat terhadap brand. Strategi retensi seperti program eksklusif, layanan prioritas, atau penawaran personalisasi dapat meningkatkan keterikatan mereka.
- Loyal Customers dan Potential Loyalists berpotensi terus bertransaksi, tetapi masih perlu insentif seperti program loyalitas berbasis poin atau rekomendasi produk yang disesuaikan untuk mencegah perpindahan ke kompetitor.
- Need Attention dan At Risk menunjukkan penurunan interaksi. Perusahaan dapat menerapkan strategi seperti diskon terbatas waktu, kampanye re-engagement, atau email pengingat yang dipersonalisasi untuk menarik mereka kembali.
Keberhasilan strategi ini dapat diukur dengan metrik seperti retention rate, repeat purchase rate, dan cost per retained customer. Dengan pendekatan berbasis data, perusahaan dapat meningkatkan loyalitas pelanggan serta mengoptimalkan anggaran pemasaran secara efisien.
