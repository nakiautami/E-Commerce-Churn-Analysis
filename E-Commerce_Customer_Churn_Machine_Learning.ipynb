## **Metric Evaluation**

**A. Memahami Biaya Churn dalam E-Commerce**

Biaya churn dalam e-commerce mencakup dua komponen utama yang harus diperhatikan untuk menjaga profitabilitas bisnis:

1. **Biaya Akuisisi Pelanggan (Customer Acquisition Cost/CAC)**: adalah biaya yang dikeluarkan untuk mendapatkan pelanggan baru, termasuk pengeluaran untuk pemasaran dan promosi. Semakin tinggi CAC, semakin besar tantangan untuk memperoleh keuntungan dari pelanggan baru. Menurut [Shopify](https://www.shopify.com/blog/customer-acquisition-cost#), untuk menghitung CAC, jumlahkan semua biaya yang terkait dengan akuisisi pelanggan baru dalam periode tertentu, kemudian bagi totalnya dengan jumlah pelanggan baru yang diperoleh dalam periode tersebut.
Rumus CAC:
$$\text{CAC} = \frac{\text{Total biaya pemasaran dan penjualan dalam periode tertentu}}{\text{Jumlah pelanggan baru yang diperoleh dalam periode tersebut}}$$ 

2. **Biaya Retensi Pelanggan (Customer Retention Cost/CRC)**: Ini adalah biaya yang diperlukan untuk mempertahankan pelanggan yang sudah ada, termasuk program loyalitas, diskon, dan biaya layanan pelanggan. Investasi dalam retensi sering kali lebih efisien dibandingkan dengan akuisisi pelanggan baru. Menurut [Forbes](https://www.forbes.com/councils/forbesbusinesscouncil/2022/12/12/customer-retention-versus-customer-acquisition/), studi menunjukkan bahwa, tergantung pada industrinya, memperoleh pelanggan baru dapat memakan biaya lima hingga tujuh kali lebih besar daripada mempertahankan pelanggan lama.
Rumus CRC:
$$\text{CRC} = \frac{\text{Total retensi dalam periode tertentu}}{\text{Jumlah pelanggan yang dipertahankan dalam periode tersebut}}$$ 

Menurut [Forbes](https://www.forbes.com/councils/forbesbusinesscouncil/2022/12/12/customer-retention-versus-customer-acquisition/), **memperoleh pelanggan baru dapat memakan biaya lima hingga tujuh kali lebih besar** daripada mempertahankan pelanggan lama.


**B. Analisis Kesalahan dalam Prediksi Churn**

Dalam analisis churn pelanggan, terdapat empat kategori dalam confusion matrix:
- **True Positive (TP)** : Customer aktualnya churn dan diprediksi churn 
- **True Negative (TN)** : Customer aktualnya tidak churn dan diprediksi tidak churn
- **False Negative (FN)** : Customer aktualnya churn dan diprediksi tidak churn
- **False Positive (FP)** : Customer aktualnya tidak churn dan diprediksi churn

Dalam analisis prediksi churn pelanggan, penting untuk memahami dua jenis kesalahan yang dapat terjadi: Type 1 error (False Positive) dan Type 2 error (False Negative).

**Type 1 error** (False Positive): Kesalahan ini terjadi ketika pelanggan yang **sebenarnya tidak churn diprediksi akan churn**. Akibatnya, perusahaan mungkin memberikan insentif seperti diskon atau promosi kepada pelanggan yang sebenarnya tetap setia, yang dapat meningkatkan loyalitas tetapi juga menambah biaya operasional.

Keuntungan:
- Pelanggan setia merasa dihargai karena mendapatkan promo atau diskon, yang dapat meningkatkan reputasi e-commerce.

Kerugian:
- Biaya promosi meningkat karena diberikan kepada pelanggan yang tidak membutuhkannya.
- Sumber daya digunakan untuk strategi retensi yang kurang tepat sasaran.

**Type 2 error** (False Negative): Kesalahan ini terjadi ketika pelanggan yang **sebenarnya churn diprediksi tidak akan churn**. Hal ini dapat menyebabkan kerugian signifikan karena perusahaan kehilangan pelanggan tanpa upaya retensi. Selain itu, perusahaan harus mengeluarkan **Customer Acquisition Cost (CAC)**, yang membutuhkan biaya besar dan waktu lebih lama, karena menarik pelanggan baru lebih sulit dibandingkan mempertahankan pelanggan lama.

Kerugian:
- Kehilangan pelanggan yang benar-benar churn.
- Meningkatnya biaya untuk memperoleh pelanggan baru sebagai pengganti.

Dalam konteks analisis churn, **Type 2 Error lebih krusial untuk dihindari karena dampaknya yang lebih besar terhadap profitabilitas perusahaan**. Ketidakmampuan mengidentifikasi pelanggan yang berisiko churn dapat menyebabkan hilangnya pelanggan dan menignkatkan CAC. Sementara itu, meskipun **Type 1 Error juga perlu diperhatikan, efeknya cenderung lebih terkait pemborosan sumber daya tetapi tidak langsung mengurangi jumlah pelanggan**.

Untuk meminimalkan kedua jenis kesalahan ini, perusahaan dapat menggunakan analisis churn yang cermat dan strategi retensi yang tepat. Pendekatan ini membantu mengidentifikasi pelanggan yang berisiko churn dan memungkinkan perusahaan mengambil tindakan proaktif untuk mempertahankan mereka, seperti menawarkan insentif yang relevan atau meningkatkan layanan pelanggan.

Selain itu, penting bagi perusahaan untuk terus memantau dan menganalisis data pelanggan guna memahami pola perilaku yang mengarah pada churn. Dengan demikian, perusahaan dapat mengembangkan strategi yang lebih efektif dalam mengurangi tingkat churn dan meningkatkan loyalitas pelanggan.

**C. Perbandingan Biaya akuisisi pelanggan (Customer Acquisition Cost/CAC)** dan **biaya retensi pelanggan (Cost Retention)**.

**1.Biaya Akuisisi Pelanggan (CAC)**: 
- Total biaya pemasaran dan penjualan untuk mendapatkan pelanggan baru.
- Contoh: Jika perusahaan menghabiskan $90.000 untuk mendapatkan 1.000 pelanggan dalam setahun, maka CAC = $90 per pelanggan.[(IBM)](https://www.ibm.com/id-id/think/topics/customer-acquisition?).

**2. Cost Retention**:
- Total biaya yang dikeluarkan untuk mempertahankan pelanggan yang sudah ada.

- Contoh: Jika perusahaan menghabiskan $100.000 untuk mempertahankan 2.500 pelanggan dalam setahun, maka CRC = $40 per pelanggan.[(ottopoint)](https://www.ottopoint.id/id/article/akuisisi-vs-retensi-pada-bisnis-mana-lebih-baik?)

Karena biaya retensi umumnya lebih rendah daripada biaya akuisisi, perusahaan perlu mengoptimalkan strategi retensi pelanggan agar dapat mengurangi churn dan meningkatkan profitabilitas jangka panjang.

**D. Mengoptimalkan Pengelolaan Churn dengan F2-Score**

Untuk mencapai keseimbangan dalam pengelolaan churn, perusahaan dapat memanfaatkan metrik F2-Score. F2-Score adalah varian dari F1-Score yang menggabungkan precision dan recall menjadi satu nilai, dengan penekanan lebih besar pada recall. Dalam analisis churn, penggunaan F2-Score sangat penting karena:

- Fokus pada Pelanggan Berisiko Churn
F2-Score lebih mengutamakan recall, sehingga perusahaan dapat lebih mudah menemukan pelanggan yang kemungkinan besar akan berhenti menggunakan layanan. Hal ini sangat penting bagi perusahaan yang ingin mempertahankan lebih banyak pelanggan dan mengurangi tingkat churn.

- Optimalisasi Sumber Daya
Dengan memahami faktor penyebab churn, perusahaan dapat mengalokasikan sumber daya secara lebih efisien dan menghindari pemborosan pada pelanggan yang tetap loyal.

- Menyeimbangkan Kesalahan Prediksi
Metrik ini membantu menyeimbangkan antara false positives (pelanggan yang diprediksi churn tetapi sebenarnya tidak) dan false negatives (pelanggan yang tidak terdeteksi akan churn tetapi sebenarnya pergi). Dengan pendekatan ini, strategi retensi dapat lebih efektif.

Rumus F2-Score:
$$F_2= \frac{(1+2^2) \times \text{Precision} \times \text{Recall}}{2^2 \times \text{Precision} \times \text{Recall}}$$

Nilai F2-Score berkisar antara 0 hingga 1, di mana nilai yang lebih tinggi menunjukkan kinerja prediksi yang lebih baik. Dengan menggunakan F2-Score, perusahaan dapat lebih memprioritaskan pencegahan churn tanpa mengabaikan akurasi prediksi.

## Uji Statistik untuk Fitur Kategorikal

Untuk memahami hubungan antara fitur kategorikal dengan variabel target (churn), kita melakukan uji statistik Chi-Square. Uji ini membantu menentukan apakah ada perbedaan signifikan antara kategori dalam fitur tertentu terhadap kemungkinan churn.

```python
import pandas as pd
import scipy.stats as stats

cat_cols = df.select_dtypes(include=['object', 'category']).columns  

target = 'Churn'

chi_results = []

for col in cat_cols:
    if col == target:
        continue 

    contingency_table = pd.crosstab(df[col], df[target])

    stat, p, dof, expected = stats.chi2_contingency(contingency_table)

    chi_results.append({
        "Kolom": col,
        "p-value": p,
        "Signifikan": "Ya" if p < 0.05 else "Tidak"
    })

chi_results_df = pd.DataFrame(chi_results)

chi_results_df
```

Hasil uji Chi-Square menunjukkan bahwa sebagian besar fitur kategorikal memiliki perbedaan yang signifikan terhadap churn, kecuali kolom Gender. Karena kolom ini tidak menunjukkan hubungan yang signifikan secara statistik, kita memutuskan untuk menghapusnya agar model tidak terganggu oleh fitur yang tidak memberikan kontribusi prediktif.

## Uji Statistik untuk Fitur Numerikal

Setelah mengevaluasi fitur kategorikal, langkah berikutnya adalah melakukan uji statistik pada fitur numerikal untuk melihat hubungan antara masing-masing fitur dengan churn. Untuk analisis ini, saya menggunakan Mann-Whitney U Test, sebuah uji non-parametrik yang cocok digunakan ketika asumsi normalitas tidak terpenuhi.

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

results = []

for col in num_cols:
    if col == 'Churn':  
        continue

    churned = df[df['Churn'] == 1][col]
    not_churned = df[df['Churn'] == 0][col]

    stat, p = stats.mannwhitneyu(churned, not_churned, alternative='two-sided')

    results.append({
        "Kolom": col,
        "p-value": p,
        "Signifikan": "Ya" if p < 0.05 else "Tidak"
    })

hasil_numerik = pd.DataFrame(results)

hasil_numerik
```

Hasil pengujian menunjukkan bahwa sebagian besar fitur memiliki perbedaan signifikan antara pelanggan churn dan non-churn. Namun, dua fitur, yaitu `HourSpendOnApp` dan `CouponUsed`, tidak menunjukkan perbedaan yang signifikan secara statistik.

Meskipun hasil uji statistik memberikan wawasan mengenai hubungan fitur dengan churn, tidak ada fitur yang dihapus pada tahap ini. Semua fitur tetap dipertahankan untuk dianalisis lebih lanjut dalam proses pemodelan machine learning.

Dalam konteks prediksi churn, pembagian data menjadi train dan test sangat krusial, karena hasil prediksi akan memengaruhi strategi retensi pelanggan serta kinerja bisnis secara keseluruhan. ## Pre-processing

### Encoding

Encoding digunakan untuk mengonversi data kategorikal ke dalam format numerik agar dapat diproses oleh algoritma machine learning. Dalam proyek ini, fitur seperti `PreferredLoginDevice`, `PreferredPaymentMode`, `PreferredOrderCat`, dan `MaritalStatus` perlu dikodekan agar dapat diintegrasikan dengan fitur numerik lainnya. Proses encoding tidak hanya memungkinkan model memahami informasi dari data kategorikal, tetapi juga berkontribusi pada peningkatan akurasi prediksi.  

Berbagai teknik encoding umum digunakan tergantung pada karakteristik data:  

- **One-Hot Encoding**: Mengubah setiap kategori dalam variabel menjadi kolom terpisah dengan nilai biner (0 atau 1). Cocok untuk data kategorikal tanpa urutan yang jelas dan jumlah kategori relatif sedikit (kurang dari 10).  
- **Ordinal Encoding**: Memberikan nilai numerik berdasarkan urutan kategori (misalnya "Rendah" = 1, "Sedang" = 2, "Tinggi" = 3). Digunakan ketika ada hubungan urutan antar kategori.  
- **Binary Encoding**: Mengonversi kategori menjadi representasi biner untuk mengurangi jumlah dimensi, berguna ketika kategori sangat banyak.  
- **Label Encoding**: Memberikan setiap kategori nilai numerik unik tanpa mempertimbangkan urutan. Cocok untuk algoritma yang tidak sensitif terhadap urutan nilai.  
- **Target Encoding**: Mengganti nilai kategori dengan rata-rata dari target variabel, misalnya dengan melihat hubungan antara kategori dan kemungkinan churn. Namun, metode ini rentan terhadap kebocoran data.  

Dalam analisis ini, digunakan **One-Hot Encoding** karena jumlah kategori dalam fitur kategorikal relatif kecil (low cardinality), sehingga lebih mudah diinterpretasikan dan menghindari masalah urutan yang mungkin muncul dalam metode seperti Label Encoding.  

Untuk menghindari multikolinearitas, diterapkan parameter `drop='first'` pada OneHotEncoder. Langkah ini memastikan bahwa satu kategori dihapus dari setiap fitur sehingga model tetap stabil dalam proses pelatihan.  

Fitur yang dikodekan dengan One-Hot Encoding dalam proyek ini meliputi:  
- `PreferredLoginDevice`  
- `PreferredPaymentMode`  
- `PreferredOrderCat`  
- `MaritalStatus`

### Scaling

Dalam proses transformasi skala fitur, digunakan **RobustScaler**. Metode ini mengukur skala berdasarkan **median** dan **interquartile range (IQR)**, yaitu selisih antara kuartil pertama (Q1) dan kuartil ketiga (Q3). Pemilihan **RobustScaler** didasarkan pada karakteristik dataset yang mengandung **outlier**, yang tetap dipertahankan dalam analisis ini. Berbeda dari metode lain yang sensitif terhadap nilai ekstrem, **RobustScaler tetap stabil** karena median dan IQR tidak terpengaruh oleh outlier.  

Selain itu, dataset ini tidak mengikuti distribusi normal, sehingga **RobustScaler lebih sesuai** karena metode ini berfokus pada pusat distribusi (median) dibandingkan dengan mean. Proses scaling dilakukan untuk memastikan bahwa data siap digunakan dalam model machine learning yang sensitif terhadap skala fitur, seperti **Logistic Regression, SVM, dan KNN**. Dengan menerapkan **RobustScaler**, skala setiap fitur menjadi lebih konsisten, meningkatkan kinerja model, serta mengurangi risiko distorsi akibat perbedaan skala antar variabel.

```python
transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), categorical_features),
    ('robust', RobustScaler(), numerical_features)
], remainder='passthrough')
```

## Resampling

Dataset yang digunakan memiliki distribusi target yang tidak seimbang (**imbalanced data**), di mana jumlah pelanggan yang melakukan churn jauh lebih sedikit dibandingkan yang tidak churn. Ketidakseimbangan ini dapat menyebabkan model machine learning cenderung **bias terhadap kelas mayoritas**, sehingga sulit untuk memprediksi kelas minoritas dengan akurat.  

Untuk mengatasi masalah ini, saya menerapkan teknik **resampling**, yaitu menyeimbangkan jumlah data antara kelas mayoritas dan minoritas dengan metode berikut:  

1. **Random OverSampling (ROS)** â†’ Menambahkan sampel baru ke kelas minoritas dengan cara menduplikasi data yang sudah ada.  
2. **SMOTE (Synthetic Minority Over-sampling Technique)** â†’ Menciptakan sampel baru secara sintetis dengan menginterpolasi titik-titik dari kelas minoritas.  
3. **ADASYN (Adaptive Synthetic Sampling)** â†’ Mirip dengan SMOTE, tetapi lebih fokus pada pembuatan sampel sintetis di area yang lebih sulit diklasifikasikan.  
4. **Random UnderSampling (RUS)** â†’ Mengurangi jumlah sampel dari kelas mayoritas secara acak untuk menyeimbangkan dataset.  
5. **NearMiss** â†’ Memilih sampel dari kelas mayoritas yang paling mirip dengan kelas minoritas untuk dihapus, sehingga distribusi data menjadi lebih seimbang.  

Dengan menerapkan teknik resampling ini, model dapat belajar dari kedua kelas secara lebih adil dan meningkatkan akurasi prediksi untuk kasus customer churn.

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler, NearMiss

ros = RandomOverSampler(random_state = 42)
smote = SMOTE(random_state = 42)
rus = RandomUnderSampler(random_state = 42)
nearmiss = NearMiss()
adasyn = ADASYN(random_state=42)
```

## Cross Validation

Cross Validation adalah teknik dalam machine learning yang digunakan untuk mengevaluasi performa model dengan lebih akurat dan mengurangi risiko overfitting. Evaluasi model sangat penting dalam machine learning karena membantu menentukan seberapa baik model dapat menggeneralisasi data baru yang belum pernah dilihat sebelumnya. Tanpa evaluasi yang tepat, model bisa saja terlihat memiliki performa tinggi pada data latihan tetapi gagal saat diuji dengan data nyata. Overfitting sendiri terjadi ketika model terlalu menyesuaikan diri dengan data pelatihan sehingga kehilangan kemampuan untuk mengenali pola pada data baru. Dengan cross validation, data dibagi menjadi beberapa bagian sehingga model diuji pada berbagai subset data yang berbeda, memastikan bahwa performanya tidak hanya bergantung pada satu set data tertentu, tetapi dapat bekerja secara konsisten pada berbagai kondisi.

```python
logreg = LogisticRegression(random_state=42)

# KNN
knn = KNeighborsClassifier()

# Decision Tree
tree = DecisionTreeClassifier(random_state=42)

# SVM
svm = SVC(random_state=42)

# Voting Classifier (Hard)
voting_hard = VotingClassifier(
    estimators=[
        ('clf1', logreg),
        ('clf2', knn),
        ('clf3', tree)
    ], voting='hard'
)

# Voting Classifier (Soft)
voting_soft = VotingClassifier(
    estimators=[
        ('clf1', logreg),
        ('clf2', knn),
        ('clf3', tree)
    ], voting='soft'
)

# Stacking Classifier
stacking = StackingClassifier(
    estimators=[
        ('clf1', logreg),
        ('clf2', knn),
        ('clf3', tree)
    ],
    final_estimator = logreg
)

# Bagging Classifier
bagging_clf = BaggingClassifier(
    estimator=logreg,
    random_state=42
)

# LightGBM
lgbm = LGBMClassifier(random_state=42)

# RandomForest
rf = RandomForestClassifier(random_state=42)

# Adaboost
adaboost = AdaBoostClassifier(random_state=42)

# Gradientboost
gradboost = GradientBoostingClassifier(random_state=42)

# ExtremeGradientBoost
xgboost = XGBClassifier(random_state = 42)
```

Salah satu jenis yang sering digunakan adalah K-Fold Cross Validation, yang membagi dataset menjadi k bagian yang sama besar, serta Stratified K-Fold, yang memastikan distribusi kelas tetap seimbang di setiap fold. Stratified K-Fold sangat berguna dalam kasus klasifikasi dengan jumlah sampel yang tidak seimbang antara kelas, karena tanpa stratifikasi, model bisa saja terlalu banyak berlatih pada kelas mayoritas dan kurang pada kelas minoritas. Dengan menggunakan cross validation, model dapat diuji secara lebih menyeluruh, memastikan bahwa performanya stabil dan tidak hanya bekerja baik pada data tertentu, tetapi juga mampu menangani data baru dengan lebih baik.

```python
# define algo

list_algo = [logreg, tree, knn, 
            voting_hard, voting_soft, stacking, bagging_clf, 
            lgbm, rf, adaboost, gradboost, xgboost]

list_resampler = [None, ros, rus, smote, nearmiss, adasyn]

f2_score = make_scorer(fbeta_score, beta = 2)

list_algo_name = []
list_resampler_name = []
list_f2_all = []
list_mean_f2 = []
list_std_f2 = []

skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for algo in list_algo:
    for resampler in list_resampler:
    # define algo
        model = algo

        Pipe_model = Pipeline([
        ('preprocessing', transformer),
        ('resampler', resampler),
        ('modeling', model)
        ])

    #cross validation
        cv_score = cross_val_score(
            estimator = Pipe_model,   
            X = X_train,
            y = y_train,
            scoring = f2_score,
            cv = skfold
    )

    #scoring 
        list_algo_name.append(str(algo))  
        list_resampler_name.append(str(resampler)) 
        list_f2_all.append(cv_score.round(2).tolist())  
        list_mean_f2.append(cv_score.mean())   
        list_std_f2.append(cv_score.std()) 
```

```python
df_cv = pd.DataFrame({
    "algo": list_algo_name,
    "resampler": list_resampler_name,
    "f2": list_f2_all,
    "f2_mean": list_mean_f2,
    "f2_std": list_std_f2
})
df_cv.sort_values(by='f2_mean', ascending=False)
```

Setelah menerapkan teknik **cross-validation**, berikut hasil evaluasi model dengan berbagai algoritma dan metode resampling. Metode evaluasi menggunakan **F2-score**, yang lebih menekankan pada recall dibanding precisionâ€”sesuai dengan tujuan mendeteksi churn dengan minim false negatives.

## Hyperparameter Tuning

Setelah melakukan cross-validation, kita menemukan bahwa **XGBoost + RandomOverSampler** memiliki performa terbaik berdasarkan skor F2-Score. Namun, model masih bisa ditingkatkan dengan Hyperparameter Tuning, yaitu proses mencari kombinasi parameter terbaik agar model bekerja lebih optimal.

**Kenapa Hyperparameter Tuning Diperlukan?**

Setiap model machine learning memiliki hyperparameter, yaitu parameter yang tidak dipelajari oleh model secara otomatis, tetapi harus diatur secara manual. Jika kita memilih hyperparameter yang tepat, model bisa memberikan hasil yang lebih akurat dan lebih baik dalam memprediksi customer churn.

**Metode yang Digunakan**

Kita akan menggunakan Grid Search, yaitu metode yang mencoba semua kombinasi hyperparameter yang telah ditentukan. Meskipun prosesnya memakan waktu, Grid Search menjamin bahwa kita menemukan kombinasi terbaik dalam rentang parameter yang kita tentukan.

**Hyperparameter yang Dicoba**
Beberapa hyperparameter penting yang diuji pada model **XGBoost**:

- **n_estimators**: Jumlah pohon dalam model (100, 300, 500, 1000)
- **learning_rate**: Kecepatan belajar (0.01 - 0.2)
- **max_depth**: Kedalaman maksimum pohon (3 - 8)
- **subsample**: Fraksi data yang digunakan dalam training (0.6 - 1)
- **colsample_bytree**: Fraksi fitur yang digunakan dalam setiap pohon (0.6 - 1)
- **gamma**: Threshold untuk pruning pohon (0 - 1)
- **reg_lambda**: Regularisasi L2 untuk mengurangi overfitting (0, 1, 10)
- **reg_alpha**: Regularisasi L1 untuk mencegah model terlalu kompleks (0, 1, 10)
- **min_child_weight**: Kontrol jumlah minimum sampel per leaf (1 - 10)

```python
hyperparam = {
    'modeling__n_estimators': [100, 300, 500],
    'modeling__learning_rate': [0.01, 0.05, 0.1, 0.15],
    'modeling__max_depth': [3, 4, 6],
    'modeling__subsample': [0.6, 0.8, 1],
    'modeling__colsample_bytree': [0.6, 0.8, 1],
    'modeling__gamma': [0, 0.01, 0.1, 0.5, 1],
    'modeling__reg_lambda': [0, 1, 10],
    'modeling__reg_alpha': [0, 1, 10],
    'modeling__min_child_weight': [1, 3, 5, 10]
}

Pipe_model = Pipeline([
    ('preprocessing',transformer),
    ('resampler', ros),
    ('modeling', xgboost)
    ])

skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

gridsearch = GridSearchCV(
    estimator = Pipe_model, 
    param_grid = hyperparam, 
    cv = skfold,
    scoring = f2_score,
    verbose =2,
    n_jobs = -1 
)

randomsearch = RandomizedSearchCV(
    estimator = Pipe_model,
    cv = skfold,
    param_distributions = hyperparam,
    scoring = f2_score,
    n_jobs = -1,
    random_state = 0,
    n_iter = 1000
)

gridsearch
```

```python
gridsearch.fit(X_train, y_train)
```

```python
gridsearch.best_params_
```

Dari hasil tuning, kita mendapatkan parameter yang meningkatkan performa model dibandingkan dengan default parameters.

- **n_estimators=500** â†’ Model menggunakan 500 pohon keputusan, menandakan lebih banyak iterasi untuk meningkatkan prediksi.

- **learning_rate=0.15** â†’ Nilai pembelajaran yang lebih tinggi mempercepat konvergensi model.

- **max_depth=6** â†’ Kedalaman maksimal pohon keputusan dibatasi hingga 6, mencegah overfitting.

- **gamma=0** â†’ Tidak ada pemangkasan cabang pohon berdasarkan reduction loss, yang berarti model akan membangun lebih banyak node tanpa batasan tambahan.

- **subsample=1 & colsample_bytree=1 â†’** Model menggunakan seluruh sampel dan fitur saat membangun setiap pohon.

- **reg_alpha=0 & reg_lambda=1** â†’ Regularisasi L2 (Ridge) diterapkan untuk mengurangi kompleksitas model dan membantu dalam seleksi fitur.

# Model Evaluation

## Deskripsi Best Model

Model terbaik yang diperoleh setelah hyperparameter tuning adalah **XGBoost** dengan konfigurasi optimal yang meningkatkan performa prediksi. XGBoost (Extreme Gradient Boosting) adalah algoritma ensemble learning berbasis gradient boosting yang sangat populer untuk tugas klasifikasi dan regresi. Dibandingkan dengan metode lain, XGBoost terkenal karena:

- Kecepatan dan Efisiensi â†’ Menggunakan optimasi berbasis cache dan paralelisasi untuk mempercepat pelatihan.
- Akurasi Tinggi â†’ Dapat menangkap pola kompleks dengan baik.
- Regularisasi yang Baik â†’ Memiliki teknik seperti L1 (Lasso) dan L2 (Ridge) untuk menghindari overfitting.
- Kemampuan Menangani Data yang Tidak Seimbang â†’ Dapat dikombinasikan dengan teknik resampling seperti SMOTE atau RandomOverSampler.

XGBoost adalah bagian dari metode gradient boosting, yang bekerja dengan membangun banyak pohon keputusan (decision trees) secara bertahap. Berikut langkah-langkahnya:

1. **Inisialisasi Model Awal**

- Model pertama dibuat dengan prediksi awal, biasanya rata-rata nilai target atau tebakan sederhana lainnya.
- Residual error dihitung untuk melihat seberapa jauh prediksi awal dari nilai sebenarnya.

2. **Membangun Pohon Secara Bertahap**

- Pohon keputusan baru dibuat untuk memperbaiki kesalahan dari model sebelumnya.
- Pohon ini fokus pada residual error, sehingga setiap iterasi model menjadi lebih akurat.
- Gradient descent digunakan untuk mengarahkan pembelajaran menuju prediksi yang lebih baik.

3. **Pembobotan Pohon dan Regularisasi**

- Setiap pohon baru mendapatkan bobot (learning rate) agar tidak terlalu mendominasi model.
- Parameter **gamma, lambda, dan alpha** digunakan untuk mengontrol kompleksitas pohon agar tidak overfitting.

4. **Prediksi Final**

- Setelah beberapa iterasi, semua pohon digabungkan untuk menghasilkan prediksi akhir dengan cara menjumlahkan output dari setiap pohon.

## Classification Report

**Classification Report** adalah alat evaluasi dalam machine learning yang digunakan untuk menilai kinerja model klasifikasi. Laporan ini memberikan gambaran mendetail mengenai bagaimana model memprediksi setiap kelas dalam dataset.  

### **Manfaat Classification Report**  
- Membantu dalam mengevaluasi model secara menyeluruh, sehingga memungkinkan pengembang untuk memahami kinerja model terhadap berbagai kelas.  
- Mengidentifikasi apakah model bekerja lebih baik pada satu kelas dibandingkan kelas lainnya, sehingga dapat membantu dalam memilih atau menyempurnakan model.  

### **Elemen-Elemen dalam Classification Report**  

1. **Precision**  
   - Mengukur akurasi prediksi positif yang benar dibandingkan dengan total prediksi positif.  
   - Digunakan untuk mengetahui seberapa andal model dalam mengklasifikasikan suatu kelas sebagai positif.  
   - **Rumus:**  

    $$\text{Precision} = \frac{TP}{TP + FP}$$
    

2. **Recall**  
   - Mengukur kemampuan model dalam menemukan seluruh contoh positif yang sebenarnya.  
   - Cocok digunakan dalam skenario di mana mendeteksi semua kasus positif lebih penting dibandingkan dengan meminimalkan kesalahan positif.  
   - **Rumus:**  
     
    $$\text{Recall} = \frac{TP}{TP + FN}$$

3. **F1 Score**  
   - Merupakan rata-rata harmonik dari precision dan recall.  
   - Berguna dalam kasus data yang tidak seimbang untuk memberikan keseimbangan antara precision dan recall.  
   - **Rumus:**  

    $$F1 = 2 \times \left(\frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}\right)$$
    

4. **Support**  
   - Menunjukkan jumlah contoh yang sebenarnya ada untuk setiap kelas dalam dataset.  
   - Membantu dalam memahami distribusi data antar kelas.  

5. **Accuracy**  
   - Mengukur rasio jumlah prediksi yang benar dibandingkan dengan keseluruhan prediksi.  
   - **Rumus:**  

    $$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    

### **Penjelasan dalam Analisis E-Commerce Churn**  
- **TP (True Positives):** Konsumen yang benar-benar churn dan berhasil dideteksi oleh model. Semakin tinggi TP, semakin baik model dalam mengidentifikasi pelanggan yang akan churn.  
- **TN (True Negatives):** Konsumen yang tidak churn dan terdeteksi dengan benar sebagai pelanggan setia.  
- **FP (False Positives):** Konsumen yang sebenarnya tidak churn tetapi salah diklasifikasikan sebagai churn, yang dapat menyebabkan tindakan yang tidak diperlukan, seperti pemberian promo kepada pelanggan yang sebenarnya setia.  
- **FN (False Negatives):** Konsumen yang seharusnya churn tetapi tidak terdeteksi oleh model, sehingga berpotensi menyebabkan kehilangan pelanggan tanpa adanya intervensi yang tepat.

## Precision-Recall Curve

**Precision-Recall Curve (PR-Curve)** adalah metode visualisasi yang digunakan untuk mengevaluasi kinerja model klasifikasi, terutama pada kasus **imbalanced data**. Grafik ini menunjukkan hubungan antara **precision** dan **recall** pada berbagai nilai threshold prediksi.  

### **Karakteristik PR-Curve**  
- **Model yang baik** memiliki precision dan recall yang tinggi, sehingga kurva cenderung berada di bagian kanan atas grafik.  
- Dengan menganalisis kurva ini, developer dapat menyesuaikan **threshold prediksi** sesuai dengan kebutuhan.  
  - Jika **false positives** harus diminimalkan, maka threshold yang lebih tinggi dapat dipilih untuk meningkatkan precision.  
  - Jika **false negatives** lebih berbahaya, maka threshold yang lebih rendah dapat digunakan untuk meningkatkan recall.

Dalam konteks machine learning, SHAP memungkinkan kita untuk mengukur sejauh mana setiap fitur memengaruhi keputusan yang dibuat oleh model.

```python
import shap

# Ambil model XGBoost dari pipeline
tree_model = final_model.named_steps['modeling']  # Ambil model

# Ambil hanya preprocessing (tanpa resampler) dan transform X_test
X_test_processed = final_model.named_steps['preprocessing'].transform(X_test)

# Buat explainer berdasarkan model
explainer = shap.TreeExplainer(tree_model)

# Hitung SHAP values
shap_values = explainer.shap_values(X_test_processed)

# Summary Plot untuk feature importance
shap.summary_plot(shap_values, X_test_processed, feature_names=feature_names)
```

```python
# Function to compute RFM Scores and Segment Customers
def compute_rfm(df):
    rfm_df = df.copy()
    
    # 'DaySinceLastOrder' = 'Recency' 
    # 'OrderCount' = 'Frequency' 
    # 'CashbackAmount' = 'Monetary'
    
    # Adjust bins based on actual data range
    bins_recency = [-1, 7, 14, 30, 31]
    bins_frequency = [0, 3, 7, 12, 16]
    bins_monetary = [0, 50, 150, 250, 325]
    labels = [4, 3, 2, 1]

    rfm_df['R_Score'] = pd.cut(rfm_df['DaySinceLastOrder'], bins=bins_recency, labels=labels, include_lowest=True).astype('int64')
    rfm_df['F_Score'] = pd.cut(rfm_df['OrderCount'], bins=bins_frequency, labels=labels, include_lowest=True).astype('int64')
    rfm_df['M_Score'] = pd.cut(rfm_df['CashbackAmount'], bins=bins_monetary, labels=labels, include_lowest=True).astype('int64')

    # Compute total RFM Score
    rfm_df['RFM_Score'] = rfm_df['R_Score'] + rfm_df['F_Score'] + rfm_df['M_Score']

    # Define customer segments based on RFM Score
    def segment_customer(score):
        if score >= 10:
            return "Best Customers"
        elif score >= 8:
            return "Loyal Customers"
        elif score >= 6:
            return "Potential Loyalists"
        elif score >= 4:
            return "At Risk"
        else:
            return "Churned Customers"

    rfm_df['Customer_Segment'] = rfm_df['RFM_Score'].apply(segment_customer)
    
    return rfm_df
```

    - Customer Retention Rate: Mengevaluasi apakah strategi intervensi yang dilakukan berdasarkan prediksi model benar-benar meningkatkan retensi pelanggan.
    - Keberhasilan Intervensi: Memantau apakah pelanggan yang teridentifikasi sebagai berisiko churn dan mendapatkan intervensi benar-benar bertahan lebih lama dibanding yang tidak mendapatkan intervensi.

Dengan menerapkan langkah-langkah di atas, model prediksi churn dapat menjadi alat yang efektif dalam mendukung pengambilan keputusan bisnis. Integrasi yang baik, evaluasi berkala, serta implementasi dashboard yang komprehensif akan memastikan model tidak hanya bekerja secara optimal tetapi juga mampu beradaptasi terhadap perubahan tren pelanggan. Selain itu, penerapan MLOps, interpretasi model, dan evaluasi dampak bisnis akan semakin meningkatkan keberhasilan implementasi model ini dalam strategi retensi pelanggan.
